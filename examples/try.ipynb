```python
import numpy as np
import pandas as pd

# Create DataFrame and list
df = pd.DataFrame(data=np.random.randint(low=0, high=4, size=10), columns=['target'])
labels = ['one', 'two', 'three']

# Function to map integers to strings using labels list
def map_to_labels(value):
    if value < len(labels):
        return labels[value]
    else:
        return None  # Handle out-of-bounds indices

# Apply function to create a new column 'label' with mapped strings
df['label'] = df['target'].apply(map_to_labels)

print(df)
```
